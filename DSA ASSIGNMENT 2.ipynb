# Read the dataset 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
df=pd.read_csv('cars_data.csv')
```

```python
df
```

# null values present in the dataset

```python
df.isnull()
```

# bar graph of male vs female buyers 

```python
 df_male=df[df["Buyer Gender"]== 'Male']
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
df_female=df[df["Buyer Gender"]== 'Female']
gender=['Male','Female'];
count=[len(df_male),len(df_female)];
ax.bar(gender,count)
plt.show()
```

# top 5 cars based on their sales price

```python
top_cars=df.sort_values(
...     by="Sale Price",
...     ascending=False)
top_cars.head(5)
```

# least 5 cars based on their Resell price.


```python
least_resale_value_cars=df.sort_values(
...     by="Resell Price",
...     ascending=True)
least_resale_value_cars.head(5)
```
